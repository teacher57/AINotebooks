```python
# 1 (без номера)
# Полином первой степени

from numpy import *
x = array([3, 1, -1, -3])
y = array([-0.5, 1.2, 4.4, 7])

A = vstack([x, ones(len(x))]).T
A
```

```python
m, c = linalg.lstsq(A, y, rcond = None)[0]
print(m, c)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(1, 1, 1)

plt.plot(x, y, 'o', markersize=10)
plt.plot(x, m * x + c, 'r')

ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.savefig("1st.png")
plt.show()
```

```python
# Полином второй степени

delta = 4
x = linspace(-10, 10, 11)
print(x)
y = x ** 2 + delta * (random.rand(11) - 0.5)
x += delta * (random.rand(11) - 0.5)
x.tofile("x_data.txt", '\n')
y.tofile("y_data.txt", '\n')
```

```python
x = fromfile("x_data.txt", float, sep='\n')
y = fromfile("y_data.txt", float, sep='\n')
print(x)
print(y)
```

```python
m = vstack((x ** 2, x, ones(11))).T
s = linalg.lstsq(m, y, rcond = None)[0]

x_prec = linspace(-10, 10, 101)
print(x_prec)

fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(1, 1, 1)

ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.plot(x, y, 'o')

plt.plot(x_prec, s[0] * x_prec ** 2 + s[1] * x_prec + s[2], '-', lw=2)
plt.grid()
plt.savefig("2nd.png")
```

```python
# Полином третьей степени

for i in range(len(y)):
    y[i] = y[i] * 10
    if x[i] > 0:
        y[i] = y[i] * -1

m = vstack((x ** 3, x ** 2, x, ones(11))).T
s = linalg.lstsq(m, y, rcond = None)[0]

x_prec = linspace(-10, 10, 101)
print(x_prec)

fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(1, 1, 1)

ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.plot(x, y, 'o')

plt.plot(x_prec, s[0] * x_prec ** 3 + s[1] * x_prec ** 2 + s[2] * x_prec + x[3], '-', lw=2)
plt.grid()
plt.savefig("3rd.png")
```

```python
# 2 (без номера)

from scipy.optimize import curve_fit
beta = (0.25, 0.75, 0.5)

def f(x, b0, b1, b2):
    return b0 + b1 * exp(-b1 * x ** 2)

x = linspace(0, 5, 51)
y_orig = f(x, *beta) 
y = y_orig + 0.05 * random.randn(len(x))
print(x, y, sep='\n')

beta_opt, beta_cov = curve_fit(f, x, y)
print(beta_opt)

lin_dev = sum(beta_cov[0])
print(lin_dev)

residuals = y - f(x, *beta_opt)
fres = sum(residuals ** 2)
print(fres)
```

```python
fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(1, 1, 1)
ax.scatter(x, y)
ax.plot(x, y_orig, 'r', lw=2)
ax.plot(x, f(x, *beta_opt), 'g', lw=2)

ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.show()
```

```python
# 3 (без номера)

from pandas import read_csv, DataFrame
url = "https://raw.githubusercontent.com/AnnaShestova/salary-years-simple-linear-regression/master/Salary_Data.csv"
data = read_csv(url)
plt.plot(data["YearsExperience"], data["Salary"], "o")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x = data.iloc[:, :-1].values
y = data.iloc[:, 1].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
regressor = LinearRegression()
regressor.fit(x_train, y_train)
print(regressor.intercept_)
print(regressor.coef_)
```

```python
y_pred = regressor.predict(x_test)
df = DataFrame({"Actual" : y_test, "Predicted" : y_pred})
df
```

```python
fig = plt.figure(figsize=(11, 6))
plt.subplot(1, 2, 2)
plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred, c='r', lw=1)

plt.subplot(1, 2, 1)
plt.scatter(x, y)
plt.plot(x_test, y_pred, c='r', lw=1)
```

```python
# 4 (без номера)

url = "https://raw.githubusercontent.com/aniruddhachoudhury/Red-Wine-Quality/master/winequality-red.csv"
data = read_csv(url)
data
```

```python
x = array(data[["alcohol", "residual sugar", "density"]])
x = DataFrame(x).rename(columns = {0 : "x1", 1 : "x2", 2 : "x3"})
y = array(data[["quality"]])[:, 0]
y = DataFrame(y).rename(columns = {0 : "y"})
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=0)
regressor = LinearRegression()
regressor.fit(x_train, y_train)
coeff = DataFrame(regressor.coef_[0], x.columns, columns=["Coefficient"])
coeff
```

```python
y_pred = regressor.predict(x_test).T[0]
y_test = array(y_test).T[0]
df = DataFrame({"Actual" : y_test, "Predicted" : y_pred})
df
```

```python
from sklearn.metrics import mean_squared_error
print("MSE =", mean_squared_error(y_test, y_pred))
```
