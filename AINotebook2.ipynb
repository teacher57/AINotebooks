```python
# problem 1.3.1
import numpy as np
a = np.zeros(64).reshape(8, 8)
for i in range(8):
    for j in range(8):
        if (i + j) % 2 == 1:
            a[i][j] = 1
print(a)
```

```python
# problem 1.3.2
a = np.arange(0, 5 * 5).reshape(5, 5)
for i in range(1, 5):
    for j in range(5):
        a[i][j] -= i * 5
print(a)
```

```python
# problem 1.3.3
a = np.random.random((5, 5, 5))
print(a)
```

```python
# problem 1.3.4

# size x*y
x = 10
y = 10

a = np.zeros((x, y))
for i in range(x):
    for j in range(y):
        if i == 0 or i == y - 1 or j == 0 or j == x - 1:
            a[i][j] = 1
print(a)
```

```python
# problem 1.3.5
a = np.random.random(10)
print("original array:\n", a)
print("\nsorted array:\n", np.sort(a))
print("\nsorted backwards:\n", np.sort(a)[::-1])
```

```python
# problem 1.3.6

# size x*y*z
x = 2
y = 3
z = 4

a = np.zeros((x, y, z))

print(a)
print(np.shape(a))
print(np.size(a))
print(np.ndim(a))
```
